Model Building

```python
class NN_3_Hidden_Layers:
  # Constructor class
    def __init__(self, input_size, hidden_size_1, hidden_size_2, hidden_size_3, output_size):
       # Parameter initialisation
        self.input_size = input_size
        self.hidden_size_1 = hidden_size_1
        self.hidden_size_2 = hidden_size_2
        self.hidden_size_3 = hidden_size_3
        self.output_size = output_size

        self.initialize_weights()

    def initialize_weights(self):

        self.W1 = np.random.randn(self.input_size, self.hidden_size_1) * 0.01
        self.b1 = np.zeros((1, self.hidden_size_1))

        self.W2 = np.random.randn(self.hidden_size_1, self.hidden_size_2) * 0.01
        self.b2 = np.zeros((1, self.hidden_size_2))

        self.W3 = np.random.randn(self.hidden_size_2, self.hidden_size_3) * 0.01
        self.b3 = np.zeros((1, self.hidden_size_3))

        self.W4 = np.random.randn(self.hidden_size_3, self.output_size) * 0.01
        self.b4 = np.zeros((1, self.output_size))

    def sigmoid(self, Z):
        return 1 / (1 + np.exp(-Z))

    def sigmoid_derivative(self, A):
        return A * (1 - A)

    def forward_propagation(self, X):
        #For 3 hidden layers
        #Aggregation
        Z1 = np.dot(X, self.W1) + self.b1
        #Activation
        A1 = self.sigmoid(Z1)

        Z2 = np.dot(A1, self.W2) + self.b2
        A2 = self.sigmoid(Z2)

        Z3 = np.dot(A2, self.W3) + self.b3
        A3 = self.sigmoid(Z3)

        # Output layer
        Z4 = np.dot(A3, self.W4) + self.b4
        A4 = self.sigmoid(Z4)

        return A1, A2, A3, A4

    def compute_cost(self, A4, Y):
        m = Y.shape[0]
        cost = -(1 / m) * np.sum(Y * np.log(A4 + 1e-10) + (1 - Y) * np.log(1 - A4 + 1e-10))
        return cost

    def back_propagation(self, X, Y, A1, A2, A3, A4):
        m = X.shape[0]

        # Gradients for output layer
        dZ4 = A4 - Y
        dW4 = (1 / m) * np.dot(A3.T, dZ4)
        db4 = (1 / m) * np.sum(dZ4, axis=0, keepdims=True)

        # Gradients for 3rd hidden layer
        dA3 = np.dot(dZ4, self.W4.T)
        dZ3 = dA3 * self.sigmoid_derivative(A3)
        dW3 = (1 / m) * np.dot(A2.T, dZ3)
        db3 = (1 / m) * np.sum(dZ3, axis=0, keepdims=True)

        # Gradients for 2nd hidden layer
        dA2 = np.dot(dZ3, self.W3.T)
        dZ2 = dA2 * self.sigmoid_derivative(A2)
        dW2 = (1 / m) * np.dot(A1.T, dZ2)
        db2 = (1 / m) * np.sum(dZ2, axis=0, keepdims=True)

        # Gradients for 1st hidden layer
        dA1 = np.dot(dZ2, self.W2.T)
        dZ1 = dA1 * self.sigmoid_derivative(A1)
        dW1 = (1 / m) * np.dot(X.T, dZ1)
        db1 = (1 / m) * np.sum(dZ1, axis=0, keepdims=True)

        return dW1, db1, dW2, db2, dW3, db3, dW4, db4

    def update_parameters(self, dW1, db1, dW2, db2, dW3, db3, dW4, db4, learning_rate):
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W3 -= learning_rate * dW3
        self.b3 -= learning_rate * db3
        self.W4 -= learning_rate * dW4
        self.b4 -= learning_rate * db4

    def train(self, X, Y, epochs, learning_rate):
        costs = []
        for epoch in range(epochs):
            # Forward propogation
            A1, A2, A3, A4 = self.forward_propagation(X)
            cost = self.compute_cost(A4, Y)
            costs.append(cost)

            # Backward propogation
            dW1, db1, dW2, db2, dW3, db3, dW4, db4 = self.back_propagation(X, Y, A1, A2, A3, A4)

            # Updating weights
            self.update_parameters(dW1, db1, dW2, db2, dW3, db3, dW4, db4, learning_rate)

            if epoch % 20 == 0:
                print(f'Epoch {epoch}, Cost: {cost}')

        plt.plot(costs)
        plt.xlabel('Epochs')
        plt.ylabel('Cost')
        plt.title('Cost vs. Epochs')
        plt.show()

    def predict(self, X):
        _, _, _, A4 = self.forward_propagation(X)
        return (A4 > 0.5).astype(int)
```

Model Training

```python
input_size = X_train.shape[1]
hidden_size_1 = 10
hidden_size_2 = 10
hidden_size_3 = 10
output_size = 1
learning_rate = 0.01
epochs = 1000

# Instance of class
nn = NN_3_Hidden_Layers(input_size, hidden_size_1, hidden_size_2, hidden_size_3, output_size)
# Train function
nn.train(X_train, y_train, epochs, learning_rate)

```

Model Accuracy

```python
y_pred = nn.predict(X_test)
accuracy = np.mean(y_pred == y_test) * 100
print(f'Test Accuracy: {accuracy:.2f}%')
```
